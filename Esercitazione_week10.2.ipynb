## DATI DIABETES

```python
diabetes = pd.read_csv("Databases/diabetes.csv")
```

```python
diabetes.head(5)
```

```python
diabetes.describe()
```

```python
# restituisco per ogni fascia d'età.
# Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?

filtro_10 = diabetes["Age (years)"] < 20
filtro_20 = (20 <= diabetes["Age (years)"]) & (diabetes["Age (years)"] <= 29)   # metto <= nella prima parentesi perchè voglio fare il 'compreso' -> MA devo mettere una '&'
filtro_30 = (30 <= diabetes["Age (years)"]) & (diabetes["Age (years)"] <= 39)
filtro_40 = (40 <= diabetes["Age (years)"]) & (diabetes["Age (years)"] <= 49)
filtro_50 = diabetes["Age (years)"] >= 50
dpb10 = diabetes.loc[filtro_10, "Diastolic blood pressure (mm Hg)"].mean()
dpb20 = diabetes.loc[filtro_20, "Diastolic blood pressure (mm Hg)"].mean()
dpb30 = diabetes.loc[filtro_30, "Diastolic blood pressure (mm Hg)"].mean()
dpb40 = diabetes.loc[filtro_40, "Diastolic blood pressure (mm Hg)"].mean()
dpb50 = diabetes.loc[filtro_50, "Diastolic blood pressure (mm Hg)"].mean()
print(dpb10, dpb20, dpb30, dpb40, dpb50)
```

```python
# filtro per ogni età e restituisco una media di "Diastolic blood pressure (mm Hg)" e di...  e voglio 'media' e 'mediana', per ogni singolo anno di età.

filtro_diabete = diabetes.groupby("Age (years)")[["Diastolic blood pressure (mm Hg)","2-Hour serum insulin (mu U/ml)"]].agg(["mean","median"])

display(filtro_diabete)
```

## DATI POKEMON

```python
poke = pd.read_csv("Databases/pokemon.csv")
poke.shape
```

```python
# creazione dell'indicizzazione
poke.set_index("#")
```

```python
poke.info()
```

```python
# Quali sono i Pokémon leggendari?

filtro_legend = poke["Legendary"]
print(poke[filtro_legend].value_counts(), "\n")    # in automatico prende solo i T
filtro_grass = poke["Type 1"] == "Grass"
filtro = filtro_legend & filtro_grass
print(f"Questi sono i leggendari di tipo 1 Grass. \n {display(poke.loc[filtro])}")
```

```python
# I leggendari di tipo 1 Ice o Fire?

filtro_legend = poke["Legendary"]
filtro_ice = poke["Type 1"] == "Ice"
filtro_fire = poke["Type 1"] == "Fire"
filtro = filtro_legend & (filtro_ice | filtro_fire)   # se non avessi le parentesi, avrebbe più peso l'&.
poke.loc[filtro]
```

```python
# Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?

filtro_gen = poke["Generation"] == 1
filtro_atk = poke["Attack"] > 50
filtro_hp = poke["HP"] < 60
filtro = filtro_gen & filtro_atk & filtro_hp
poke.loc[filtro]
```
