```python
# quante attività ci sono nel quadrato di longitudine 9-10 e latitudine 45-46?

filtro_lon = (9 <= pub["Longitudine"]) & (pub["Longitudine"] <= 10)
filtro_lat = (45 <= pub["Latitudine"]) & (pub["Latitudine"] <= 46)
filtro = filtro_lon & filtro_lat
print(pub.loc[filtro].shape)
print("Il numero di attività tra la logitudine 9-10 e latitudine 45-46 sono:", pub.loc[filtro].shape[0])  # indico solo le righe                                         
```

```python
# quante attività ci sono nella provincia di Vicenza?

filtro_vic = pub["Provincia"] == "VICENZA"
print("Possiamo osservare le informazioni sulle attività situate a Vicenza\n", pub.loc[filtro_vic])

attivita_vic = pub.groupby(filtro_vic)["Comune"].count()    # voglio sapere quanti ci sono.
print(f"\nNella provincia di Vicenza ci sono n° attività attive: {attivita_vic}")   

#pub[filtro_vic].shape[0]           #posso fare anche così, facendomi restituire solo il numero delle righe 
```

```python
# quante enoteche ci sono, e come si chiamano?

enoteche = []

for nome in pub["Nome"]:
  if type(nome) != float:         # voglio che lo faccia sui valori 'esistenti', e che non mi prenda i valori 'nulli'.
    nome_min = nome.lower()       # se ci sono dati 'nullli' da dei problemi
    if "enoteca" in nome_min:
        enoteche.append(nome)
       
print(enoteche)
print("Ci sono", len(enoteche), "enoteche")
```
